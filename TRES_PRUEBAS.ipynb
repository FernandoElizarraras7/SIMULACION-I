```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = [0.78, 0.98, 0.24, 0.73, 0.43, 0.16, 0.78, 0.47, 0.18, 0.55,
     0.04, 0.29, 0.68, 0.77, 0.16, 0.03, 0.79, 0.22, 0.37, 0.80,
     0.96, 0.26, 0.91, 0.55, 0.75, 0.55, 0.64, 0.39, 0.53, 0.45,
     0.61, 0.14, 0.38, 0.12, 0.40, 0.74, 0.78, 0.98, 0.27, 0.60,
     0.43, 0.67, 0.62, 0.32, 0.53, 0.54, 0.24, 0.29, 0.18, 0.08,
     0.82, 0.94, 0.19, 0.98, 0.41, 1.00, 0.74, 0.92, 0.14, 0.43,
     0.83, 0.88, 0.18, 0.21, 0.50, 0.13, 0.43, 0.69, 0.08, 0.12,
     0.22, 0.50, 0.16, 0.11, 0.18, 0.89, 0.80, 0.42, 0.29, 0.87,
     0.83, 0.79, 0.65, 0.28, 0.78, 0.49, 0.36, 0.86, 0.87, 0.64,
     0.51, 0.07, 0.18, 0.94, 0.50, 0.22, 0.66, 0.91, 0.48, 0.24]

#Realizamos la prueba de Chi-cuadrado
print("Las hipotesis son:")
print("\nH0: La lista de datos sigue una distribución uniforme.")
print("\nH1: La lista de datos no sigue una distribución uniforme.")

valor_critico = 9.49
print("\nValor crítico:", valor_critico)

#Contadores para intervalos
frecuencias = [0, 0, 0, 0, 0]

#Clasificar los valores en intervalos y contar frecuencias
for i in x:
    if i < 0.20:
        frecuencias[0] += 1
    elif i < 0.40:
        frecuencias[1] += 1
    elif i < 0.60:
        frecuencias[2] += 1
    elif i < 0.80:
        frecuencias[3] += 1
    else:
        frecuencias[4] += 1

#Calcular esperanza teórica
esperanza = len(x) / 5

#Calcular estadístico de prueba
estadistico_prueba = sum([(frecuencia - esperanza) ** 2 for frecuencia in frecuencias]) / esperanza
print("Estadístico de prueba:", estadistico_prueba)

#Verificar si se rechaza la hipótesis nula
if estadistico_prueba > valor_critico:
  print("Se rechaza la hipotesis nula. No tiene distribución uniforme.")
else:
  print("Se acepta la hipotesis nula. Si tiene distribución uniforme.")

#Etiquetas para los intervalos
etiquetas = ["0.01-0.20", "0.21-0.40", "0.41-0.60", "0.61-0.80", "0.81-1.00"]

#Gráfico de barras
plt.bar(etiquetas, frecuencias, color = 'darkblue')
plt.title('Distribución de la lista')
plt.xlabel('Intervalos')
plt.ylabel('Frecuencia')
plt.savefig('distribucion.png')
plt.show()

#Realizamos la prueba de Kolmogorov-Smirnov
print("Las hipótesis son:")
print("H0: La muestra sigue una distribución uniforme.")
print("H1: La muestra no sigue una distribución uniforme.")

#Se realiza la prueba Kolmogorov-Smirnov con el uso de la librería
resultado = stats.kstest(x, 'uniform')

#Obtener el valor crítico para un nivel de significancia del 5%
n = len(x)
alpha = 0.05
valor_critico = 0.136 / (n ** 0.5)

print("\nEl valor crítico es:", valor_critico)
print("El valor del estadístico de prueba es:", resultado.statistic)

#Comparar el valor del estadístico de prueba con el valor crítico
if resultado.statistic < valor_critico:
    print("\nSe acepta la hipótesis nula. La muestra sigue una distribución uniforme.")
else:
    print("\nSe rechaza la hipótesis nula. La muestra no sigue una distribución uniforme.")

#Obtener la distribución uniforme para graficar
uniforme = np.linspace(min(x), max(x), len(x))

#Calcular la distribución empírica acumulada (CDF) de los datos
ecdf = np.arange(1, len(x) + 1) / len(x)

#Graficar la distribución empírica acumulada (CDF) y la distribución uniforme
plt.plot(np.sort(x), ecdf, label='Empírica', color = 'teal')
plt.plot(uniforme, np.arange(1, len(uniforme) + 1) / len(uniforme), label='Uniforme', color = 'salmon')

#Configuraciones de la gráfica
plt.xlabel('Valor')
plt.ylabel('Probabilidad acumulada')
plt.title('Distribución Empírica vs. Distribución Uniforme')
plt.legend()
plt.grid(True)
plt.show()

#Realizamos la prueba de Corridas
#Cálculo del conjunto S
def calcular_conjunto_S(muestra):
    conjunto_S = []
    for i in range(1, len(muestra)):
        if muestra[i] > muestra[i-1]:
            conjunto_S.append(1)
        else:
            conjunto_S.append(0)
    return conjunto_S

conjunto_S = calcular_conjunto_S(x)

#Cálculo de las corridas en la muestra
def contar_corridas(conjunto_S):
    corridas = 1
    for i in range(1, len(conjunto_S)):
        if conjunto_S[i] != conjunto_S[i-1]:
            corridas += 1
    return corridas

corridas = contar_corridas(conjunto_S)

#Cálculo del estadístico de la prueba de corridas
def calcular_estadistico_corridas(muestra):
    n = len(x)
    Mc = (2 * n - 1) / 3
    var = (16 * n - 29) / 90
    conjunto_S = calcular_conjunto_S(muestra)
    corridas_observadas = contar_corridas(conjunto_S)
    Z0 = (corridas_observadas - Mc) / var
    return Z0

estadistico_corridas = calcular_estadistico_corridas(x)

#Definición del intervalo de aceptación
intervalo_aceptacion = (-1.96,1.96)

#Impresión de resultados
print("Número total de elementos en la muestra:", len(x))
print("\nLas hipótesis son:")
print("H0: Los números de la muestra son independientes")
print("H1: Los números de la muestra no son independientes")
print("\nConjunto S:", conjunto_S)
print("Número de corridas en la muestra:", corridas)
print("\nEl intervalo de aceptación es:", intervalo_aceptacion)
print(f"\nEstadístico de la prueba de corridas: {estadistico_corridas:.4f}")
print(f"-1.96 <= {estadistico_corridas:.4f} <= 1.96")

#Evaluación de la hipótesis nula
if -1.96 < estadistico_corridas < 1.96:
    print("Se acepta H0. Los números de la muestra son independientes")
else:
    print("Se rechaza H0. Los números de la muestra no son independientes")
```
