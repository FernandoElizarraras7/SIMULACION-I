# MÉTODO DE LA TRANSFORMADA INVERSA

El método de la transformada inversa es una técnica utilizada para generar muestras aleatorias a partir de una distribución de probabilidad dada. Este método es particularmente útil para distribuciones continuas
### Definir la función de densidad de probabilidad (fdp) de la distribución exponencial
La función de densidad de probabilidad de una variable aleatoria exponencial con parámetro $λ$ está dada por:
$$f(x)=λe^{-λx}$$
### GRÁFICA DE LA FDP

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer el estilo de los gráficos
sns.set(style="whitegrid")
def exponencial_fdp(x, lam):
    return lam * np.exp(-lam * x)

# Parámetro de la distribución
lam = 1.0

# Valores de x
x = np.linspace(0, 10, 1000)

# Valores de la fdp
y = exponencial_fdp(x, lam)

# Gráfica de la fdp
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'$\lambda = {lam}$', color='orange')
plt.title('Función de Densidad de Probabilidad (fdp)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Definir la función de distribución acumulativa (FDA) de la distribución exponencial

La función de distribución acumulativa de una variable aleatoria exponencial con parámetro $λ$ está dada por:
$$F(x)=1-e^{λx}$$
### Gráfica de la FDA

```python
def exponencial_fda(x, lam):
    return 1 - np.exp(-lam * x)
# Valores de la FDA
y_cdf = exponencial_fda(x, lam)

# Gráfica de la FDA
plt.figure(figsize=(10, 6))
plt.plot(x, y_cdf, label=f'$\lambda = {lam}$', color='sienna')
plt.title('Función de Distribución Acumulativa (FDA)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Generar muestras aleatorias usando el método de la transformada inversa
Para generar una muestra aleatoria $X$ de una distribución exponencial usando el método de la transformada inversa, primero generamos una muestra uniforme $U$ en el intervalo $[0,1)$. Luego, usamos la inversa de la FDA para obtener $X$:
$$X=F^{-1}(U)$$
La inversa de la FDA de una distribución exponencial con parámetro $λ$ es:
$$F^{-1}(u)=-\frac{1}{λ}ln(1-u)$$
### Generar muestras y graficar el histograma

```python
def inversa_fda_exponencial(u, lam):
    return -np.log(1 - u) / lam
# Número de muestras
num_samples = 10000

# Generar muestras uniformes
u_samples = np.random.uniform(0, 1, num_samples)

# Transformar las muestras usando la inversa de la FDA
x_samples = inversa_fda_exponencial(u_samples, lam)

# Graficar el histograma de las muestras
plt.figure(figsize=(10, 6))
plt.hist(x_samples, bins=50, density=True, alpha=0.6, color='salmon', label='Histograma de muestras')

# Graficar la fdp teórica
plt.plot(x, y, label='fdp teórica', color='rebeccapurple')
plt.title('Histograma de Muestras y fdp Teórica')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
