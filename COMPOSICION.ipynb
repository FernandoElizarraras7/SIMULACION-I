# MÉTODO DE COMPOSICIÓN
El método de composición es una técnica utilizada para generar muestras aleatorias a partir de una distribución de probabilidad que puede ser expresada como una mezcla de otras distribuciones. En este ejemplo, implementaremos el método de composición utilizando una mezcla de dos distribuciones normales.
### Definir la función de densidad de probabilidad (fdp) de la mezcla de distribuciones normales
La fdp de una mezcla de dos distribuciones normales está dada por:
$$f(x)=p⋅f_1(x)+(1-p)⋅f_2(x)$$
donde $f_1(x)$ y $f_2(x)$ son las fdp de las dos distribuciones normales y $p$ es la probabilidad de elegir la primera distribución.
### Gráfica de la fdp

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Establecer el estilo de los gráficos
sns.set(style="whitegrid")

def mezcla_fdp(x, p, mu1, sigma1, mu2, sigma2):
    return p * norm.pdf(x, mu1, sigma1) + (1 - p) * norm.pdf(x, mu2, sigma2)

# Parámetros de la mezcla
p = 0.3
mu1, sigma1 = -2, 0.5
mu2, sigma2 = 2, 1.0

# Valores de x
x = np.linspace(-6, 6, 1000)

# Valores de la fdp
y = mezcla_fdp(x, p, mu1, sigma1, mu2, sigma2)

# Gráfica de la fdp
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Mezcla de Normales\n$p={p}$, $\mu_1={mu1}$, $\sigma_1={sigma1}$, $\mu_2={mu2}$, $\sigma_2={sigma2}$', color='darkgoldenrod')
plt.title('Función de Densidad de Probabilidad (fdp)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Definir la función de distribución acumulativa (FDA) de la mezcla de distribuciones normales
La FDA de una mezcla de dos distribuciones normales está dada por:
$$F(x)=p⋅F_1(x)+(1-p)⋅F_2(x)$$
donde $F_1(x)$ y $F_2(x)$ son las FDA de las dos distribuciones normales.
### Gráfica de la FDA

```python
def mezcla_fda(x, p, mu1, sigma1, mu2, sigma2):
    return p * norm.cdf(x, mu1, sigma1) + (1 - p) * norm.cdf(x, mu2, sigma2)
# Valores de la FDA
y_cdf = mezcla_fda(x, p, mu1, sigma1, mu2, sigma2)

# Gráfica de la FDA
plt.figure(figsize=(10, 6))
plt.plot(x, y_cdf, label=f'Mezcla de Normales\n$p={p}$, $\mu_1={mu1}$, $\sigma_1={sigma1}$, $\mu_2={mu2}$, $\sigma_2={sigma2}$', color='magenta')
plt.title('Función de Distribución Acumulativa (FDA)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### Generar muestras aleatorias usando el método de composición
Para generar muestras de una mezcla de distribuciones, seguimos estos pasos:


1.   Elegir una de las distribuciones con probabilidad $p$ y $1−p$.
2.   Generar una muestra de la distribución elegida.




```python
def generar_muestras_mezcla(n, p, mu1, sigma1, mu2, sigma2):
    muestras = []
    for _ in range(n):
        if np.random.rand() < p:
            muestras.append(np.random.normal(mu1, sigma1))
        else:
            muestras.append(np.random.normal(mu2, sigma2))
    return np.array(muestras)

# Número de muestras
num_samples = 10000
muestras = generar_muestras_mezcla(num_samples, p, mu1, sigma1, mu2, sigma2)

# Graficar el histograma de las muestras
plt.figure(figsize=(10, 6))
plt.hist(muestras, bins=50, density=True, alpha=0.6, color='crimson', label='Histograma de muestras')

# Graficar la fdp teórica
plt.plot(x, y, label='fdp teórica', color='chartreuse')
plt.title('Histograma de Muestras y fdp Teórica')
plt.xlabel('x')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
