# Estimación de π mediante el Método de Monte Carlo
El método de Monte Carlo es una técnica numérica que utiliza números aleatorios para resolver problemas matemáticos. En el contexto de la estimación de π, aprovechamos la relación entre el área de un círculo y un cuadrado que lo contiene para obtener una aproximación del valor de π.
### Descripción del Método


1.   **Área del Círculo y el Cuadrado:** Consideremos un círculo de radio $r$ inscrito en un cuadrado de lado $2r$. El área del círculo es $πr^2$
  y el área del cuadrado es $(2r)^2=4r^2$
2.   **Relación de Áreas:** La probabilidad de que un punto aleatorio dentro del cuadrado caiga dentro del círculo es proporcional al área del círculo respecto al área del cuadrado:
$$\frac{Area\,del\,círculo}{Area\,del\,cuadrado}=\frac{πr^2}{4r^2}=\frac{π}{4}$$
3. **Estimación de π:** Si generamos $N$ puntos aleatorios uniformemente distribuidos dentro del cuadrado y contamos cuántos caen dentro del círculo, entonces la fracción de puntos dentro del círculo debería aproximarse a $\frac{π}{4}$
4. **Fórmula de Estimación:** Podemos estimar π como:
$$π≈4⋅\left (\frac{Numero\,de\,puntos\,dentro\,del\,circulo}{Numero\,total\,de\,puntos\,generados}  \right )$$

## EJERCICIO
Se van a generar puntos aleatorios dentro de un cuadrado unitario y contar cuántos de estos puntos caen dentro de un círculo unitario inscrito en el cuadrado. Basándonos en la proporción de puntos dentro del círculo con respecto al total de puntos generados, estimaremos el valor de π.
### Pasos del Ejercicio


1.   **Generación de Puntos Aleatorios:** Generaremos $N$ puntos aleatorios uniformemente distribuidos dentro de un cuadrado de lado 2 centrado en el origen (cuadrado unitario).
2.   **Cálculo de Puntos dentro del Círculo:** Verificaremos para cada punto generado si está dentro del círculo unitario utilizando la condición $x^2+y^2≤1$, donde $(x,y)$ son las coordenadas del punto.
3. **Estimación de π:** Usando la proporción de puntos dentro del círculo con respecto al total de puntos generados, estimaremos π según la fórmula:
$$π≈4⋅\left (\frac{Numero\,de\,puntos\,dentro\,del\,circulo}{Numero\,total\,de\,puntos\,generados}  \right )$$
4. **Implementación en Python:** Se va a implementar este proceso utilizando Python para obtener una estimación de π con $N=10000$ puntos.

## IMPLEMENTACIÓN EN PYTHON

```python
import numpy as np
import matplotlib.pyplot as plt

# Función para estimar pi usando el método de Monte Carlo
def estimar_pi_monte_carlo(n):
    puntos_dentro = 0

    for _ in range(n):
        x = np.random.uniform(-1, 1)
        y = np.random.uniform(-1, 1)

        # Verificar si el punto (x, y) está dentro del círculo unitario
        if x**2 + y**2 <= 1:
            puntos_dentro += 1

    # Calcular la estimación de pi
    pi_estimado = 4 * puntos_dentro / n
    return pi_estimado

# Parámetros del experimento
N = 10000  # Número de puntos a generar

# Estimar pi usando Monte Carlo
pi_estimado = estimar_pi_monte_carlo(N)

# Mostrar resultado
print(f"Estimación de pi con {N} puntos: {pi_estimado}")
```

## EXPLICACIÓN DEL CÓDIGO


*   **Generación de Puntos Aleatorios:** Se generan $N$ puntos aleatorios en el rango $[−1,1]$ para $x$ e $y$, que definen el cuadrado unitario centrado en el origen.
*   **Cálculo de Puntos dentro del Círculo:** Para cada punto generado, se verifica si está dentro del círculo unitario usando la condición $x^2+y^2≤1$.
*   Estimación de π: Se calcula $π$ multiplicando por 4 la fracción de puntos que caen dentro del círculo respecto al total de puntos generados.
*   Visualización del Resultado: El resultado se imprime en la consola con la estimación de π basada en los $N$ puntos generados.
## VISUALIZACIÓN GRÁFICA



```python
# Graficar los puntos generados
x_inside = []
y_inside = []
x_outside = []
y_outside = []

for _ in range(N):
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)

    if x**2 + y**2 <= 1:
        x_inside.append(x)
        y_inside.append(y)
    else:
        x_outside.append(x)
        y_outside.append(y)

plt.figure(figsize=(8, 8))
plt.scatter(x_inside, y_inside, color='darkcyan', s=5, label='Puntos dentro del círculo')
plt.scatter(x_outside, y_outside, color='darkred', s=5, label='Puntos fuera del círculo')
plt.title(f'Estimación de π con Monte Carlo: {pi_estimado:.5f}')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
