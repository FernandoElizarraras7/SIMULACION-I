```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

lista_numeros = [0.8797, 0.3884, 0.6289, 0.8750, 0.5999, 0.8589, 0.9996, 0.2415, 0.3808, 0.9606,
                 0.9848, 0.3469, 0.7977, 0.5844, 0.8147, 0.6431, 0.7387, 0.5613, 0.0318, 0.7401,
                 0.4557, 0.1592, 0.8536, 0.8846, 0.3410, 0.1492, 0.8681, 0.5291, 0.3188, 0.5992,
                 0.9170, 0.2204, 0.5991, 0.5461, 0.5739, 0.3254, 0.0856, 0.2258, 0.4603, 0.5027,
                 0.8376, 0.6235, 0.3681, 0.2088, 0.1525, 0.2006, 0.4720, 0.4272, 0.6360, 0.0954]

print("Las hipótesis son:")
print("H0: La muestra sigue una distribución uniforme.")
print("H1: La muestra no sigue una distribución uniforme.")

#Se realiza la prueba Kolmogorov-Smirnov con el uso de la librería
resultado = stats.kstest(lista_numeros, 'uniform')

#Obtener el valor crítico para un nivel de significancia del 5%
n = len(lista_numeros)
alpha = 0.05
valor_critico = 1.36 / (n ** 0.5)

print("\nEl valor crítico es:", valor_critico)
print("El valor del estadístico de prueba es:", resultado.statistic)

#Comparar el valor del estadístico de prueba con el valor crítico
if resultado.statistic < valor_critico:
    print("\nSe acepta la hipótesis nula. La muestra sigue una distribución uniforme.")
else:
    print("\nSe rechaza la hipótesis nula. La muestra no sigue una distribución uniforme.")

#Obtener la distribución uniforme para graficar
uniforme = np.linspace(min(lista_numeros), max(lista_numeros), len(lista_numeros))

#Calcular la distribución empírica acumulada (CDF) de los datos
ecdf = np.arange(1, len(lista_numeros) + 1) / len(lista_numeros)

#Graficar la distribución empírica acumulada (CDF) y la distribución uniforme
plt.plot(np.sort(lista_numeros), ecdf, label='Empírica', color = 'purple')
plt.plot(uniforme, np.arange(1, len(uniforme) + 1) / len(uniforme), label='Uniforme', color = 'red')

#Configuraciones de la gráfica
plt.xlabel('Valor')
plt.ylabel('Probabilidad acumulada')
plt.title('Distribución Empírica vs. Distribución Uniforme')
plt.legend()
plt.grid(True)
plt.show()
```
